DATA VISUALIZATION AND PREPROCESSSING


CHANGE M AND B TO BINARY VALUES , CHNGE ID COLUMN TO INDEX OF DATAFRAME

```python
dataset['diagnosis'] = dataset['diagnosis'].apply(lambda x: '1' if x== 'M' else '0')
```

```python
X = dataset.drop(['id', 'diagnosis'], axis=1).values
y = dataset['diagnosis'].values
```

```python
print(X)
```

```python
print(y)
```

SPLIT DATASET

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=21)
```

```python
print(X)
```

```python
print(y)
```

STANDARIZE

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X)
```

```python
print(y)
```

TRAIN DATASET

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

PREDICTION

```python
print(classifier.predict(sc.transform([[17.99, 10.38, 122.8, 1001, 0.1184, 0.2776, 0.3001, 0.1471, 0.2419, 0.07871,
           1.095, 0.9053, 8.589, 153.4, 0.006399, 0.04904, 0.05373, 0.01587, 0.03003, 0.006193,
           25.38, 17.33, 184.6, 2019, 0.1622, 0.6656, 0.7119, 0.2654, 0.4601, 0.1189]])))
```

```python
sample = [[17.99, 10.38, 122.8, 1001, 0.1184, 0.2776, 0.3001, 0.1471, 0.2419, 0.07871,
           1.095, 0.9053, 8.589, 153.4, 0.006399, 0.04904, 0.05373, 0.01587, 0.03003, 0.006193,
           25.38, 17.33, 184.6, 2019, 0.1622, 0.6656, 0.7119, 0.2654, 0.4601, 0.1189]]

scaled_sample = sc.transform(sample)
prediction = classifier.predict(scaled_sample)
print("Prediction:", prediction)
```

PREDCITION COMPARISON

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

COMPUTE ACCURACY WITH K-FOLD CROSS VALIDATION

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_test, y_pred, normalize=False)
```
